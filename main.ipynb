## Comparaison avec optimisation et validation croisée

```python
'''
Si l'on souahaite allouer plus de ressources pour créer des modèles, 

Paramètres: 
- niter : nombre de configurations d'hyperparamètres à tester
- cv : argument k de stratify k fold. Ce paramètre affecte à la fois la phase d'optimisation des hyperparmètres et la phase d'évaluation du modèles.    
- select k best & sequential_forward_search : Permet de spécifier si l'on souhaite filtrer les features extraites. 
les modèles seront déclinés avec une version SANS filtre et AVEC Filtre


- Cette méthode permet de réaliser 3 expériences. 
    1. Calcul de la RMSE pour les modèles issus des données extraites de TSFEL.
    2. Calcul de la RMSE pour les modèles issus des données extraites de TSFRESH.
    3. Calcul de la RMSE pour les modèles issus des données originaux.

'''

exp = Experience(regresseurs,output_dir,data_path,pb)
exp.changer_config_general({
            'normalisation': 'minmax',
            'train_size': 0.7,
            'cv': 5,
            'scorer': 'neg_root_mean_squared_error',
            'optimisation': 'randomized_search',
            'extraction_features': True,
            'librairie_extraction_features': 'tsfel',
            'n_iter':3,
            'sauvegarder_boxplot': True,
            'sequential_forward_search': False,
            'select_k_best': False 

        })
exp.run()
```
